```python
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
from numpy.linalg import norm
```

```python
data = pd.read_csv('checkins.dat', sep='|', skipinitialspace=True, header=0).replace(r'^\s+$', np.nan, regex=True).dropna()
```

```python
cordinates = data[['latitude      ', 'longitude     ']][:100000]
```

```python
model = MeanShift(bandwidth=0.1,bin_seeding=True, min_bin_freq=15)
preds = model.fit_predict(cordinates)
```
